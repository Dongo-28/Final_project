```python
try:
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    print("✅ Bibliotecas carregadas com sucesso!")
except ImportError as e:
    print(f"❌ Erro: {e}\nExecute no terminal: pip install seaborn pandas matplotlib")

# Carregar dados com tratamento de erros
try:
    temp_global = pd.read_csv('../data/global_temperature.csv')
    nivel_mar = pd.read_csv('../data/sea_level.csv')
    print("\n✅ Dados carregados:\n", temp_global.head())
except FileNotFoundError:
    print("❌ Erro: Verifique se os arquivos estão em '../data/' com os nomes corretos")

# Pré-processamento seguro
if 'temp_global' in locals():
    temp_global.fillna(temp_global.mean(), inplace=True)
    
    # Plot com estilo profissional
    plt.figure(figsize=(12, 6))
    sns.set_style("whitegrid")  # Estilo do seaborn
    sns.lineplot(data=temp_global, x='Year', y='Temperature', 
                 color='firebrick', linewidth=2.5, label='Temperatura Global')
    plt.title('Tendência de Temperatura Global (1880-2025)', pad=20, fontsize=14)
    plt.xlabel('Ano', fontsize=12)
    plt.ylabel('Variação (°C)', fontsize=12)
    plt.legend()
    
    try:
        plt.savefig('../reports/temp_trend.png', dpi=300, bbox_inches='tight')
        print("\n✅ Gráfico salvo em '../reports/temp_trend.png'")
    except:
        print("\n⚠️ Erro ao salvar gráfico. Crie a pasta 'reports' primeiro.")
    
    plt.show()
```
