```python
# %% [markdown]
## 1. CARREGAMENTO DE DADOS
# %% [code]
import pandas as pd
import matplotlib.pyplot as plt

# Carregar dados
temp_global = pd.read_csv('../data/global_temperature.csv')  # Caminho relativo
nivel_mar = pd.read_csv('../data/sea_level.csv')

# Verificar
print("Dados de Temperatura:")
display(temp_global.head())
```

```python
# %% [markdown]
## 2. PRIMEIRA VISUALIZAÇÃO
# %% [code]
plt.figure(figsize=(12, 6))
plt.plot(temp_global['Year'], temp_global['Temperature'], 
         color='red', linewidth=2, label='Temperatura Global')
plt.title('Variação da Temperatura Global (1880-2023)', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Variação (°C)', fontsize=12)
plt.grid(alpha=0.3)
plt.legend()
plt.savefig('../reports/temp_trend.png', dpi=300, bbox_inches='tight')  # Salva automaticamente
plt.show()
```
